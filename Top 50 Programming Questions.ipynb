# 1. Merge Strings Alternately

Question 1: You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1.
If a string is longer than the other, append the additional letters onto the end of the merged string. Return the merged string.

```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        m = len(word1)
        n = len(word2)
        i = 0
        j = 0
        result = []

        while i < m or j < n:
            if i < m:
                result += word1[i]
                i += 1
            if j < n:
                result += word2[j]
                j += 1

        return "".join(result)

if __name__ == "__main__":
    word1 = input("Enter the first word: ")
    word2 = input("Enter the second word: ")

    solution = Solution()

    merged_word = solution.mergeAlternately(word1, word2)
    print("Merged Word:", merged_word)
```

# 2. Find the Difference
Question 2: You are given two strings s and t. String t is generated by random shuffling string s and then add one more letter at a random position.
Return the letter that was added to t.

```python
class Solution:
    def findTheDifference(self, s, t):
        # Use XOR to find the differing character
        result = 0
        for char in s + t:
            result ^= ord(char)
        return chr(result)

# Interactive Part
if __name__ == "__main__":
    # Input strings
    s = input("Enter string s: ")
    t = input("Enter string t: ")

    # Create a Solution object
    solution = Solution()

    # Find and print the added letter
    added_letter = solution.findTheDifference(s, t)
    print("The added letter is:", added_letter)
```

# 3. Find the Index of the First Occurrence in a String
Question 3: Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

```python
class Solution:
    def strStr(self, haystack, needle):
        # Check if needle is empty
        if not needle:
            return 0
        
        # Loop through haystack to find the needle
        for i in range(len(haystack) - len(needle) + 1):
            # Check if substring matches needle
            if haystack[i:i+len(needle)] == needle:
                return i
        
        # If not found, return -1
        return -1

# Interactive Part
if __name__ == "__main__":
    # Input strings
    haystack = input("Enter the haystack: ")
    needle = input("Enter the needle: ")

    # Create a Solution object
    solution = Solution()

    # Find and print the result
    result = solution.strStr(haystack, needle)
    print("The index of the first occurrence is:", result)
```

# 4. Valid Anagram
Given two strings s and t, return true if t is an anagram of s, and false otherwise.

```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False
    return sorted(s) == sorted(t)

# Example usage
if __name__ == "__main__":
    # Example inputs
    s1, t1 = "anagram", "nagaram"
    s2, t2 = "rat", "car"
    
    # Outputs
    print("Example 1 Output:", isAnagram(s1, t1))  # Output: True
    print("Example 2 Output:", isAnagram(s2, t2))  # Output: False
```

# 6. Move Zeroes
Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.
Note that you must do this in-place without making a copy of the array.

```python
def moveZeroes(nums):
    non_zero_index = 0  # Pointer for the next non-zero element
    
    # Move all non-zero elements to the front
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[non_zero_index], nums[i] = nums[i], nums[non_zero_index]
            non_zero_index += 1

# Example usage
if __name__ == "__main__":
    # Input from the user
    nums = list(map(int, input("Enter the array elements separated by space: ").split()))

    # Perform the operation
    moveZeroes(nums)
    
    # Output the result
    print("Output:", nums)
```

# 7. Plus One
You are given a large integer represented as an integer array digits, where each digits[i] is the ith digit of the integer. The digits are ordered from most significant to least significant in left-to-right order. The large integer does not contain any leading 0's.
Increment the large integer by one and return the resulting array of digits.

```python
def plusOne(digits):
    n = len(digits)
    
    # Traverse the digits from right to left
    for i in range(n - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1  # Increment the digit
            return digits  # Return as no carry-over is needed
        digits[i] = 0  # Set current digit to 0 (carry-over)
    
    # If all digits are 9, add a leading 1
    return [1] + digits

# Example usage
if __name__ == "__main__":
    # Input from the user
    digits = list(map(int, input("Enter the digits separated by space: ").split()))

    # Get the incremented result
    result = plusOne(digits)
    
    # Output the result
    print("Output:", result)
```

# 8. Sign of the Product of an Array
Implement a function signFunc(x) that returns:

1 if x is positive.
-1 if x is negative.
0 if x is equal to 0.
You are given an integer array nums. Let product be the product of all values in the array nums.

Return signFunc(product).

```python
def signFunc(x):
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0

def arraySign(nums):
    product = 1
    for num in nums:
        product *= num  # Multiply each number in the array
    
    return signFunc(product)  # Return the sign of the product

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    nums = list(map(int, input("Enter the array elements separated by space: ").split()))
    
    # Get the result
    result = arraySign(nums)
    
    # Output the result
    print("Output:", result)
```

# 9. Can Make Arithmetic Progression From Sequence
A sequence of numbers is called an arithmetic progression if the difference between any two consecutive elements is the same.
Given an array of numbers arr, return true if the array can be rearranged to form an arithmetic progression. Otherwise, return false.

```python
def canBeArithmeticProgression(arr):
    arr.sort()  # Sort the array first
    diff = arr[1] - arr[0]  # Calculate the difference between the first two elements
    
    # Check the difference for every consecutive pair in the sorted array
    for i in range(1, len(arr) - 1):
        if arr[i + 1] - arr[i] != diff:
            return False  # If the difference is not consistent, return False
    
    return True  # If the differences are consistent, return True

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    arr = list(map(int, input("Enter the array elements separated by space: ").split()))
    
    # Get the result
    result = canBeArithmeticProgression(arr)
    
    # Output the result
    print("Output:", result)
```

# 10. Monotonic Array
An array is monotonic if it is either monotone increasing or monotone decreasing.
An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is monotone decreasing if for all i <= j, nums[i] >= nums[j].
Given an integer array nums, return true if the given array is monotonic, or false otherwise.

```python
def isMonotonic(nums):
    # Check if the array is monotonic increasing or decreasing
    increasing = decreasing = True
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            decreasing = False
        elif nums[i] < nums[i - 1]:
            increasing = False
    
    return increasing or decreasing

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    nums = list(map(int, input("Enter the array elements separated by space: ").split()))
    
    # Get the result
    result = isMonotonic(nums)
    
    # Output the result
    print("Output:", result)
```

# 11. Roman to Integer
Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.
Given a roman numeral, convert it to an integer.

```python
def romanToInt(s):
    # Dictionary mapping Roman symbols to their integer values
    roman_map = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000
    }
    
    result = 0
    for i in range(len(s) - 1):
        # If current Roman value is less than the next one, subtract it
        if roman_map[s[i]] < roman_map[s[i + 1]]:
            result -= roman_map[s[i]]
        else:
            result += roman_map[s[i]]
    
    # Add the value of the last symbol
    result += roman_map[s[-1]]
    
    return result

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    s = input("Enter the Roman numeral: ")
    
    # Get the result
    result = romanToInt(s)
    
    # Output the result
    print("Output:", result)
```

# 14. Baseball Game
You are keeping the scores for a baseball game with strange rules. At the beginning of the game, you start with an empty record.

You are given a list of strings operations, where operations[i] is the ith operation you must apply to the record and is one of the following:

An integer x.
Record a new score of x.
'+'.
Record a new score that is the sum of the previous two scores.
'D'.
Record a new score that is the double of the previous score.
'C'.
Invalidate the previous score, removing it from the record.
Return the sum of all the scores on the record after applying all the operations.

The test cases are generated such that the answer and all intermediate calculations fit in a 32-bit integer and that all operations are valid.

```python
def calPoints(operations):
    # Initialize an empty list to store the scores
    record = []
    
    for operation in operations:
        if operation == 'C':
            # Remove the last score if 'C' is encountered
            record.pop()
        elif operation == 'D':
            # Double the last score if 'D' is encountered
            record.append(record[-1] * 2)
        elif operation == '+':
            # Add the sum of the last two scores if '+' is encountered
            record.append(record[-1] + record[-2])
        else:
            # Convert the operation to an integer and add it to the record
            record.append(int(operation))
    
    # Return the sum of all the scores in the record
    return sum(record)

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    operations = input("Enter the operations (separated by space): ").split()
    
    # Get the result
    result = calPoints(operations)
    
    # Output the result
    print("Output:", result)
```

# 15. Robot Return to Origin
There is a robot starting at the position (0, 0), the origin, on a 2D plane. Given a sequence of its moves, judge if this robot ends up at (0, 0) after it completes its moves.
You are given a string moves that represents the move sequence of the robot where moves[i] represents its ith move. Valid moves are 'R' (right), 'L' (left), 'U' (up), and 'D' (down).
Return true if the robot returns to the origin after it finishes all of its moves, or false otherwise.

Note: The way that the robot is "facing" is irrelevant. 'R' will always make the robot move to the right once, 'L' will always make it move left, etc. Also, assume that the magnitude of the robot's movement is the same for each move.

```python
def judgeCircle(moves):
    # Initialize coordinates (x, y) of the robot
    x, y = 0, 0
    
    # Iterate through each move in the moves string
    for move in moves:
        if move == 'U':
            y += 1  # Move up increases the y-coordinate
        elif move == 'D':
            y -= 1  # Move down decreases the y-coordinate
        elif move == 'L':
            x -= 1  # Move left decreases the x-coordinate
        elif move == 'R':
            x += 1  # Move right increases the x-coordinate
    
    # After all moves, check if the robot has returned to the origin (0, 0)
    return x == 0 and y == 0

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    moves = input("Enter the moves: ")
    
    # Get the result
    result = judgeCircle(moves)
    
    # Output the result
    print("Output:", result)
```

# 16. Find Winner on a Tic Tac Toe Game
Tic-tac-toe is played by two players A and B on a 3 x 3 grid. The rules of Tic-Tac-Toe are:

Players take turns placing characters into empty squares ' '.
The first player A always places 'X' characters, while the second player B always places 'O' characters.
'X' and 'O' characters are always placed into empty squares, never on filled ones.
The game ends when there are three of the same (non-empty) character filling any row, column, or diagonal.
The game also ends if all squares are non-empty.
No more moves can be played if the game is over.
Given a 2D integer array moves where moves[i] = [rowi, coli] indicates that the ith move will be played on grid[rowi][coli]. return the winner of the game if it exists (A or B). In case the game ends in a draw return "Draw". If there are still movements to play return "Pending".

You can assume that moves is valid (i.e., it follows the rules of Tic-Tac-Toe), the grid is initially empty, and A will play first.

```python
def tictactoe(moves):
    # Initialize the 3x3 board
    board = [[' ' for _ in range(3)] for _ in range(3)]

    # Function to check if there is a winner
    def check_winner():
        # Check rows and columns
        for i in range(3):
            if board[i][0] == board[i][1] == board[i][2] != ' ':
                return board[i][0]
            if board[0][i] == board[1][i] == board[2][i] != ' ':
                return board[0][i]
        
        # Check diagonals
        if board[0][0] == board[1][1] == board[2][2] != ' ':
            return board[0][0]
        if board[0][2] == board[1][1] == board[2][0] != ' ':
            return board[0][2]
        
        return None

    # Simulate the game
    for i, move in enumerate(moves):
        row, col = move
        player = 'X' if i % 2 == 0 else 'O'  # Player A (X) and Player B (O)
        board[row][col] = player
        
        # Check for a winner after each move
        winner = check_winner()
        if winner:
            return "A" if winner == 'X' else "B"

    # If all moves are done but no winner, check if there are empty spots
    if len(moves) == 9:
        return "Draw"
    
    return "Pending"

# Example usage with input
if __name__ == "__main__":
    # Input from the user
    moves = [[int(x) for x in input("Enter move (row col): ").split()] for _ in range(int(input("How many moves? ")))]
    
    # Get the result
    result = tictactoe(moves)
    
    # Output the result
    print("Output:", result)
```

# Robot Bounded In Circle
On an infinite plane, a robot initially stands at (0, 0) and faces north. Note that:

The north direction is the positive direction of the y-axis.
The south direction is the negative direction of the y-axis.
The east direction is the positive direction of the x-axis.
The west direction is the negative direction of the x-axis.
The robot can receive one of three instructions:

"G": go straight 1 unit.
"L": turn 90 degrees to the left (i.e., anti-clockwise direction).
"R": turn 90 degrees to the right (i.e., clockwise direction).
The robot performs the instructions given in order, and repeats them forever.

Return true if and only if there exists a circle in the plane such that the robot never leaves the circle.

```python
def isRobotBounded(instructions: str) -> bool:
    # Directions: North, East, South, West
    # (dx, dy) for each direction
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    # Initial position (0, 0) and initial direction (facing north)
    x, y = 0, 0
    direction = 0  # 0 -> North, 1 -> East, 2 -> South, 3 -> West
    
    for instruction in instructions:
        if instruction == 'G':
            # Move in the current direction
            dx, dy = directions[direction]
            x += dx
            y += dy
        elif instruction == 'L':
            # Turn left (counterclockwise), so decrease direction index
            direction = (direction - 1) % 4
        elif instruction == 'R':
            # Turn right (clockwise), so increase direction index
            direction = (direction + 1) % 4
    
    # If the robot is back at the origin or not facing north, it's bounded
    return (x == 0 and y == 0) or direction != 0

# Example usage:
instructions = "GGLLGG"
print(isRobotBounded(instructions))  # Output: True

instructions = "GG"
print(isRobotBounded(instructions))  # Output: False

instructions = "GL"
print(isRobotBounded(instructions))  # Output: True
```
